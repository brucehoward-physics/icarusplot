```python
import matplotlib.pyplot as plt

from icarusplot import *
```

```python
pot = 3.0e20
```

```python
fig, axs = RatioCanvas('Ratio Canvas')
axs[0].hist([1,1,2,1,2],histtype='step')

UnityLine(axs[1])

axs[0].set_ylabel('Counts')
axs[1].set_ylabel('Ratio')
axs[1].set_xlabel('Variable')

ICARUSMCPreliminary(axs[0])
POT(axs[0],pot)
```

```python
fig, axs = RatioCanvas()
axs[0].hist([0,1,1,2,1,2],bins=np.linspace(0,3,4),histtype='step')
axs[0].set_xlim(0,3)
axs[0].set_ylim(0,3.3)

UnityLine(axs[1])

axs[0].set_ylabel('Counts')
axs[1].set_ylabel('Ratio')
axs[1].set_xlabel('Variable')

ICARUSMCPreliminary(axs[0])
POT(axs[0],pot)
```
